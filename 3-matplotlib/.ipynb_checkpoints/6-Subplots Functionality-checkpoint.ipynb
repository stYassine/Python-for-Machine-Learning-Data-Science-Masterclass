- In theory we could create a Figure object and then manually add and arrange sets of axes to line up multiple plots side by side.
- However, Matplotlib comes with a pre-configured function call plt.subplots() that automatically does this for us!
<br><br>
- The plt.subplots() call allows us to easily create Figure and axes objects in the side by side formations.
- The plt.subplots() command returns a tuple containing the Figure canvas and then a numpy array holding the axes objects.
<br><br>

#### - fig, axes = plt.subplots(nrows=1, ncols=2)
<img src="./assets/6/1-sub_plots 1.png" height="100" />
<br><br>

#### - fig, axes = plt.subplots(nrows=2, ncols=2)
<img src="./assets/6/1-sub_plots 2.png" height="100" />

- plt.subplots() returns a tuple which by command convention we label (fig, axes):
- Fig: This is the entire Figure canvas.
- Axes: this is the a numpy array holding each of the axes according to position in the overall canvas.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
```

```python
a
```

```python
b = a**4
```

```python
x = np.arange(0, 10)
```

```python
x
```

```python
y = 2 * x
```

```python
y
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[1].plot(a, b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[1][1].plot(a, b)
```

```python
print(axes.shape)
```

```python
type(axes)
```

```python
axes
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

for ax in axes:
    ax.plot(x, y)
```

```python
# Add spacing 1
fig, axes = plt.subplots(nrows=3, ncols=1)

fig.tight_layout()

for ax in axes:
    ax.plot(x, y)
```

```python
# Add spacing 2
fig, axes = plt.subplots(nrows=2, ncols=2)

fig.subplots_adjust(wspace=0.5, hspace=0.5)

for ax in axes:
    for item in ax:
        item.plot(x, y)
```

```python
# Parameters
fig, axes = plt.subplots(nrows=2, ncols=2)
# you can pass parames like this

# plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

axes[0][0].plot(a, b)
axes[0][0].set_ylabel('Y Label 0,0')
axes[0][1].plot(x, y)

axes[1][0].plot(a, b)
axes[1][0].set_title('Title 1,0')
# .set_xlim(2, 6)
axes[1][1].plot(x, y)

# Super Level Title
fig.suptitle('Super Title Level', fontsize=18)

# 
fig.set_figwidth(10)

fig.subplots_adjust(hspace=1, wspace=0.4)
```
