```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
```

```python
a
```

```python
b = a**4
```

```python
b
```

```python
x = np.arange(0, 10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
fig = plt.figure() # blank figure

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, y)

fig.show() # hide the message above the figure
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(a, b)
```

```python
# Multiple Axes
fig = plt.figure()

# Large Axes
axes_1 = fig.add_axes([0,0,1,1])
axes_1.plot(a, b)
axes_1.set_xlim(0,8)
axes_1.set_ylim(0,8000)
axes_1.set_xlabel('A')
axes_1.set_ylabel('B')
axes_1.set_title('Power of 4')


# Small Axes
axes_2 = fig.add_axes([0.2,0.5, 0.25, 0.25])
axes_2.plot(x, y)
axes_2.set_xlim(1,2)
axes_2.set_ylim(0,10)
axes_2.set_xlabel('A')
axes_2.set_ylabel('B')
axes_2.set_title('Zommed In')

# fig.show()
```

```python
# Figure Parameters
# dpi: the more value the more clear the figure is (less blurry)
# figsize: plot size in inches
fig = plt.figure(figsize=(4, 4), dpi=100)

axes = fig.add_axes([0, 0, 1, 1])

axes.plot(a, b)

# include x & y axes when saving the plot
fig.savefig('new_figure.png', bbox_inches='tight')
```
