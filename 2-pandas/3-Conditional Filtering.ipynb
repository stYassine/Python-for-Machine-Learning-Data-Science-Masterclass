- Typically in data Analysis our datasets are large enough that we don't filter based on position, but instead based on a condition
- Conditional filterin allows us to select rows based a cdition on a column.
- This leads to a discussion on organizing our data...

## Filter by single condition

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./tips.csv')
```

```python
df.head()
```

### Single Condition

```python
bool_series = df['total_bill'] > 40
```

```python
df[bool_series]
```

```python
male_series = df['sex'] == 'Male'
```

```python
df[male_series]
```

```python
df[df['size'] > 3]
```

## Filter by multiple condition

```python
# AND & --- Both Condition need to be True
# OR | --- Either Condition is True
```

```python
bill_cond = df['total_bill'] > 30
sex_cond = df['sex'] == 'Male'
```

```python
### AND
# df[bill_cond & sex_cond]
# or like this
df[(df['total_bill'] > 30) & (df['sex'] == 'Male')]
```

```python
### OR
df[(df['total_bill'] > 30) | (df['sex'] == 'Male')]
```

```python
## Check against multiple possible values
```

```python
df[(df['day'] == 'Fri') | (df['day'] == 'sat') | (df['day'] == 'sun')]
```

```python
# In
days = ['Sat', 'Sun']
df[df['day'].isin(days)]
```
