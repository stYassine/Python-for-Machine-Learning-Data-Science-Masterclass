- Matrix plots are the visual equivalent of displaying a pivot table.
- The matrix plot displaying all the data passed in visualizing all the numeric values in a DataFrame.
#### Note!
- Not every DataFrame is a valid choice for a matrix plot such as a heatmap.
<br><br>
The two main matrix plot types are:
- heatmap(): Visually displays the distribution of cell values with a color mapping.
- clustermap(): Same visual as heatmap, but first conducts hierarchical clustring to reorganize data into groups.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./country_table.csv')
```

```python
df.head()
```

```python
# RESET THE INDEX
df = df.set_index('Countries')
```

```python
# This is a Good Data for heatmap
df.head()
```

```python
sns.heatmap(df)
```

```python
sns.heatmap(df.drop('Life expectancy', axis=1))
```

```python
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5)
```

```python
plt.figure(dpi=100)
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True)
```

```python
plt.figure(dpi=100)
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis')
```

```python
plt.figure(dpi=100)
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis', center=True)
```

#### Clustermap

```python
plt.figure(dpi=100)
sns.clustermap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis', center=True)
```

```python
plt.figure(dpi=100)
sns.clustermap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis', center=True, col_cluster=False)
```
