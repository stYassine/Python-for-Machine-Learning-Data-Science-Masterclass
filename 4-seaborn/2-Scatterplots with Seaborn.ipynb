- Scatter plots show the relationship between two continuous features.
- Recall that <b>Continuous</b> features are numeric variables that can take any number of values between any two values.
<br><br>
- Continuous Features Examples:
	- Age
	- Height
	- Salary
	- Temperature
	- Prices
<br><br>
- A Continuous features allows for a value to always be between two values.
- Not to be confused with <b>Categorical</b> features which represent distinct and unique categories:
	- Colors
	- Shapes
	- Names
<br><br>

- Scatter plots line up a set of two continuous features and plots them out as coordinates.
- For example, imagine employyes with salaries who sell a certain dollar amout of items each year.
We could explore the relationship between exmployee salaries and sales amount.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df, hue='level of education')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df, hue='salary')
```

```python
# Add different colors (Pallets)
plt.figure(figsize=(10, 6), dpi=100)
sns.scatterplot(x='salary',y='sales',data=df, hue='level of education', palette='Dark2')
```

```python
# points size
plt.figure(figsize=(10, 6), dpi=100)
sns.scatterplot(x='salary',y='sales',data=df, size='work experience')
```

```python
# All Points size
plt.figure(figsize=(10, 4), dpi=100)
sns.scatterplot(x='salary',y='sales',data=df, s=100)
```

```python
# Transparency
# from 0.0 to 1
plt.figure(figsize=(10, 4), dpi=100)
sns.scatterplot(x='salary',y='sales',data=df, s=100, alpha=0.1)
```

```python
# Marker Style
plt.figure(figsize=(10, 4), dpi=100)
sns.scatterplot(x='salary',y='sales',data=df, s=100, alpha=1, hue='level of education', style='level of education')

# Save image
plt.savefig('my_plot.jpg')
```
