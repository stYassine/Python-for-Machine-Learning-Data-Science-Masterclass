```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

### Boxplot

```python
sns.boxplot(y='math score',data=df)
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.boxplot(y='math score', x='test preparation course', data=df)
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.boxplot(y='reading score', x='test preparation course', data=df)
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.boxplot(y='math score', x='parental level of education', data=df)
```

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.boxplot(y='math score', x='parental level of education', hue='test preparation course', data=df)
plt.legend(bbox_to_anchor=(1.05, .5))
```

```python
# Switch X & Y
plt.figure(figsize=(10, 4), dpi=100)
sns.boxplot(y='parental level of education', x='math score', hue='test preparation course', data=df)
plt.legend(bbox_to_anchor=(1.05, .5))
```

### ViolinPlot

```python
plt.figure(figsize=(10, 8), dpi=100)
sns.violinplot(data=df, x='reading score', y='parental level of education')
```

```python
plt.figure(figsize=(10, 8), dpi=100)
sns.violinplot(data=df, x='reading score', y='parental level of education', hue='test preparation course', palette='Set2')
plt.legend(bbox_to_anchor=(1.55, 0.5))
```

```python
plt.figure(figsize=(10, 8), dpi=100)
sns.violinplot(data=df, x='reading score', y='parental level of education', 
               hue='test preparation course', palette='Set2', split=True)
plt.legend(bbox_to_anchor=(1.55, 0.5))
```

```python
plt.figure(figsize=(10, 8), dpi=100)
sns.violinplot(data=df, x='reading score', y='parental level of education', 
               hue='test preparation course', palette='Set2', split=True, inner=None)
plt.legend(bbox_to_anchor=(1.55, 0.5))
```

```python
plt.figure(figsize=(10, 8), dpi=100)
sns.violinplot(data=df, x='reading score', y='parental level of education', 
               hue='test preparation course', split=True, inner='quartile')
plt.legend(bbox_to_anchor=(1.55, 0.5))
```

```python
plt.figure(figsize=(10, 8), dpi=100)
sns.violinplot(data=df, x='reading score', y='parental level of education', 
               hue='test preparation course', split=True, inner='stick')
plt.legend(bbox_to_anchor=(1.55, 0.5))
```

```python
plt.figure(figsize=(10, 8), dpi=100)
sns.violinplot(data=df, x='reading score', y='parental level of education', 
               hue='test preparation course', bw=0.1)
```

### Swarmplot

```python
sns.swarmplot(x='math score',data=df)
```

```python
sns.swarmplot(x='math score',data=df, size=3)
```

```python
plt.figure(figsize=(10, 8), dpi=100)
sns.swarmplot(x='math score',data=df, size=3)
```

```python
plt.figure(figsize=(10, 8), dpi=100)
sns.swarmplot(x='math score', y='gender', hue='test preparation course', data=df, size=3)

plt.legend(bbox_to_anchor=(1.35, 0.5))
```

```python
plt.figure(figsize=(10, 8), dpi=100)
sns.swarmplot(x='math score', y='gender', hue='test preparation course', data=df, size=3, dodge=True)

plt.legend(bbox_to_anchor=(1.35, 0.5))
```
