```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

### 1- RugPlot

```python
sns.rugplot(x='salary',data=df)
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
plt.figure(figsize=(5,8), dpi=100)
sns.rugplot(x='salary',data=df)
```

### 2- Displot

```python
sns.displot(x='salary',data=df)
```

```python
sns.displot(x='salary',data=df, bins=10)
```

```python
sns.displot(x='salary',data=df, bins=100)
```

```python
# This is on SNS Level Global for all plots
sns.set(style='darkgrid')
sns.displot(x='salary',data=df, bins=20)
```

```python
sns.set(style='whitegrid')
sns.displot(x='salary',data=df, bins=20)
```

```python
sns.set(style='dark')
sns.displot(x='salary',data=df, bins=20)
```

```python
sns.set(style='ticks')
sns.displot(x='salary',data=df, bins=20)
```

```python
sns.displot(x='salary',data=df, bins=20, color='red', edgecolor='blue', linewidth=2, ls='--')
```

### 3 - KDE

```python
sns.displot(x='salary',data=df, kde=True)
```

```python
# add rug
sns.displot(x='salary',data=df, kde=True, rug=True)
```

### 4 histplot

```python
sns.histplot(x='salary', data=df)
```

```python
# Kde Plot
sns.kdeplot(x='salary',data=df)
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(x='age', data=sample_ages)
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=30)
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=30, kde=True)
```

```python
sns.kdeplot(x='age', data=sample_ages)
```

```python
sns.kdeplot(x='age', data=sample_ages, clip=[0, 100])
```

```python
sns.kdeplot(x='age', data=sample_ages, clip=[0, 100], shade=True)
```

```python
# bw_adjust: Bandwith
sns.kdeplot(x='age', data=sample_ages, clip=[0, 100], shade=True, bw_adjust=0.1)
```

```python

sns.kdeplot(x='age', data=sample_ages, clip=[0, 100], shade=True, bw_adjust=1)
```
