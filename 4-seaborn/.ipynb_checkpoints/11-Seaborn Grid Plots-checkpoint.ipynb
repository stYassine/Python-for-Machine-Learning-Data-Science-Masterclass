- Seaborn grid calls use Matplotlib subplots() to automatically create a grid based off a categorical column.
- Instead of passing in a specific number of cols or rows for the subplots, we can simply pass in the name of the column and seaborn will automatically map the subplots grid.
<br><br>
- Many of seaborn's built-in plot calls are running on top of this grid system.
- Directly calling the grid system allows users to heavily customize plots.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.catplot(x='gender',y='math score',data=df, kind='box')
```

```python
# create row for every category in 'Lunch'
sns.catplot(x='gender',y='math score',data=df, kind='box', row='lunch')
```

```python
sns.catplot(x='gender',y='math score',data=df, kind='box', hue='lunch')
```

```python
# create column for every category in 'Lunch'
sns.catplot(x='gender',y='math score',data=df, kind='box', col='lunch')
```

```python
sns.catplot(x='gender',y='math score',data=df, kind='box', col='lunch', row='test preparation course')
```

### PairGrid

```python
# create based on every col
sns.PairGrid(df)
```

```python
# Upper
grid = sns.PairGrid(df)

grid.map_upper(sns.scatterplot)
```

```python
# Diagnol
grid = sns.PairGrid(df)

grid.map_diag(sns.histplot)
```

```python
# Lower
grid = sns.PairGrid(df)

grid.map_lower(sns.kdeplot)
```

```python
# All
grid = sns.PairGrid(df)

grid.map_upper(sns.scatterplot)
grid.map_diag(sns.histplot)
grid.map_lower(sns.kdeplot)
```

```python
# Hue
grid = sns.PairGrid(df, hue='gender')

grid.map_upper(sns.scatterplot)
grid.map_diag(sns.histplot)
grid.map_lower(sns.kdeplot)

grid = grid.add_legend()
```
